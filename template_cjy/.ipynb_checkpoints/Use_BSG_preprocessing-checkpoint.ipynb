```python
import warnings
warnings.filterwarnings("ignore")
from BSG_preprocessing import pack, load_RealData, denoise, unpack, process_data
from BSG_preprocessing_individual import processs_data_individual
```

```python
import numpy as np
dataset_train = np.load("../data/all_back_train.npy")
dataset_test = np.load("../data/all_back_test.npy")
print(dataset_train.shape)
print(dataset_test.shape)
```

```python
# load the real data
labels_train, labels_test, signals_train, signals_test = load_RealData(dataset_path_train='../data/all_back_train.npy',
                                                dataset_path_test='../data/all_back_test.npy')

print(signals_train.shape)

# pack the signals and labels
package_test = pack(signals_test, labels_test)

# denoise the signals and filter out bad quality signals.
# core function is from Yingjian Song
denoised_package_test = denoise(package_test)
print(denoised_package_test["signals"].shape)
```

```python
# Method 1:
# process all signals at one time
# filtered_package_train, realigned_all_pieces_train, templates_train = process_data(denoised_package_train)

# Method 2:
# process signal piece by piece
signals, labels, ids = unpack(denoised_package_test)
list_template_test = []
for cnt, the_id in enumerate(ids):
    label = labels[cnt]
    signal = signals[cnt]

    energy, good_pieces, template = processs_data_individual(signal, label)

    if template is None:
        continue
    """
    Anything ...
    """
    list_template_test.append(template)

np.save("../data/template_test.npy",list_template_test)   
```
